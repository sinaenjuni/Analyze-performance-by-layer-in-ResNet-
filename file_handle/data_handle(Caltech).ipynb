```python
import os, shutil
import errno
```

```python
def make_sure_path_exists(path):
    try:
        os.makedirs(path)
    except OSError as exception:
        if exception.errno != errno.EEXIST:
            raise
```

```python
original_dataset_dir = './Images/Caltech/101/101_ObjectCategories/'
target_dataset_dir = './Images/Caltech/101/101_images/'
```

```python
make_sure_path_exists(target_dataset_dir)
```

```python
original_list = os.listdir(original_dataset_dir)
for x in original_list:
    print(x)
print(len(original_list))
```

```python
a_name = ['train', 'test', 'validation']
b_name = original_list
```

```python
count = 0
for a in a_name:
    for b in b_name:
        make_sure_path_exists(target_dataset_dir+'/'+a+'/'+b)
        print(target_dataset_dir+'/'+a+'/'+b)
        count += 1
print(count)
```

```python
print(os.path.join(original_dataset_dir, original_list[0]))
```

```python
t = 0
index = 0
for x in original_list:
    print(x)
    list = os.listdir(os.path.join(original_dataset_dir, x))
    list.sort()
    #print(list)
    print(len(list))
    train_size = int(len(list)*0.6)
    test_val_size = int(len(list)*0.2)

    for y in range(len(list)):
        #print(list[y])
        t+=1
        shutil.copyfile(os.path.join(original_dataset_dir+'/'+x, list[y]),
                       os.path.join(target_dataset_dir+a_name[index]+'/'+x+'/'+list[y]))
        print(os.path.join(original_dataset_dir, list[y])+ ' TO ' +
                        os.path.join(target_dataset_dir+a_name[index]
                                     +'/'+x+'/'+list[y]))            
        
        #print(os.path.join(target_dataset_dir+a_name[index]+'/'+x+'/'+list[y]))
        
        if t == train_size:
            index+=1
        elif t == train_size+test_val_size:
            index+=1

            
    print('t:{}'.format(t))
    train_size=0
    test_val_size=0
    index=0
    t=0
        
```
